##ステップ関数
$
h(x) =
\begin{cases}
0, & \mbox{if } x < 0, \\
1, & \mbox{otherwise }
\end{cases}

```python
import numpy as np

def step_function(x):
  if type(x) is not int:
    print('x is not int')
  else:
    if x < 0:
      return 0
    else: return 1

x = np.array([0, 1, 2, 3, -4])
step_function(x)
```

```python
def step_function(x):
  return np.array(x > 0, dtype=int)

x = np.array([1, 2, -8, 0])
# x = 8
# np.arange(-5.0, 5.0, 0.1)
step_function(x)
```

```python
# グラフを描画する
import numpy as np
import matplotlib.pylab as plt

x = np.arange(-5, 5, 0.1) # -5から5まで0.1おきに
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # y軸の範囲を指定
plt.show()
```

##シグモイド関数
##$h(x) = \frac{1}{1 + e^{-x}}$

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

x = np.arange(-4, 4, 0.1)
y = sigmoid(x)

plt.plot(x, y)

def sigmoid_2(x):
  return np.exp(x) / (1 + np.exp(x))

z = sigmoid_2(x)
```
